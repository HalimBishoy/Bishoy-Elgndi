```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.125
fig = plt.subplots(figsize =(18, 10))
 
# set height of bar
DT = [60, 0, 0, 63.125,0,88]
NB = [58, 0, 55.91,53.43,0,89]
RF = [0, 99.5, 65.83, 100,66.54,92]
LR = [0,0,0,59.68,0,92]
KNN = [0,0,0,48.75,0,90]
SVM = [0,0,67.25,49.68,70.33,0]
 
# Set position of bar on X axis
br1 = np.arange(len(DT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
br6 = [x + barWidth for x in br5]
br7 = [x + barWidth for x in br6] 
br8 = [x + barWidth for x in br7] 
# Make the plot
plt.bar(br1, DT, color ='r', width = barWidth,
        edgecolor ='grey', label ='DT')
plt.bar(br2, NB, color ='g', width = barWidth,
        edgecolor ='grey', label ='NB')
plt.bar(br3, RF, color ='b', width = barWidth,
        edgecolor ='grey', label ='RF')
plt.bar(br4, LR, color ='c', width = barWidth,
        edgecolor ='grey', label ='LR')
plt.bar(br6, KNN, color ='k', width = barWidth,
        edgecolor ='grey', label ='KNN')
plt.bar(br4, SVM, color ='m', width = barWidth,
        edgecolor ='grey', label ='SVM')


 
# Adding Xticks
plt.xlabel('Reference Number', fontweight ='bold', fontsize = 16)
plt.ylabel('Accuracy %', fontweight ='bold', fontsize = 16)
plt.xticks([r + barWidth for r in range(len(DT))],
        ['Ref 3', 'Ref 6', 'Ref 8', 'Ref 11', 'Ref 12','Our Work'])

# Set the font size of xticks
plt.xticks(fontsize=18)
# Set the font size of yticks
plt.yticks(fontsize=18)
plt.legend(fontsize=18)
 

plt.show()
```
